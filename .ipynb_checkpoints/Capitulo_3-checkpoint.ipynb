### QAM (QUADRATURE AMPLITUDE MODULATION)
La modulación QAM hace uso de la variación de la amplitid y de la fase de la portadora

```python
import numpy as np 
import matplotlib.pyplot as plt 
import komm

fs = 1000 #frecuencia "real"
np.random.seed(40)

def x_t(simbols,tS,fc=1,separadas=False):
    t_c = np.arange(0,tS,1/fs)
    
    tf = tS*len(simbols)
    t = np.arange(0,tf,1/fs)

    x = np.cos(2*np.pi*t_c*fc)
    y = np.sin(2*np.pi*t_c*fc)
    
    if not separadas:
        fm = np.array([])
        for i in simbols:
            fm = np.append(fm, i.real*x-i.imag*y) 
        return t,fm
    
    else: 
        phase = np.array([])
        qua = np.array([])
        for i in simbols:
            phase = np.append(phase, i.real*x) 
            qua = np.append(qua,i.imag*y) 
        return phase,qua
        

def combin(bits): 
    x = np.zeros((2**bits, bits))
    for i in range(x.shape[0]):
        b_ = bin(i)[2:]
        for j in range(len(b_)):
            x[i,x.shape[1]-len(b_)+j] = int(b_[j])
    return x 

def raisedCosinePulse(roll_off,span):
    t = np.arange(-4*span,4*span,1/fs)
    a = np.cos(np.pi*roll_off*t/span)/(1-(4*roll_off**2*t**2)/span**2)
    h = np.sinc(t/span)*a
    return h 

def drawFullEyeDiagram(xt,Fs=fs,T=1):
    """Draw the eye diagram using all parts of the given signal xt
    T es el la duración del simbolo
    FS es la frecuencia de muestreo"""
    T = 0.5
    samples_perT = int(Fs*T)
    samples_perWindow = int(2*Fs*T)
    parts = []
    startInd = 3*samples_perT  # ignore some transient effects at beginning of signal
    for k in range(int(len(xt)/samples_perT)-1):
        if max(startInd + k*samples_perT + np.arange(samples_perWindow)) > (len(xt)-startInd)  :
            break
        parts.append(xt[startInd + k*samples_perT + np.arange(samples_perWindow)])
        
    parts = np.array(parts).T
    
    t_part = np.arange(-T, T, 1/Fs)
    plt.plot(t_part, parts, 'b-')
    
def eye_diagram(mod,SNR,fil):
    bits = bits = np.random.randint(0,2,int(mod.order*np.log2(mod.order))*2)
    symbols_tx = mod.modulate(bits)
    symbols_rx = komm.AWGNChannel(SNR)(symbols_tx)
    phase,qua = x_t(symbols_rx,1,separadas=True)
    x_phase = np.convolve(phase,fil,mode="same")
    x_qua = np.convolve(qua,fil,mode="same")
    plt.figure(figsize=(20,5))
    plt.subplot(1,2,1)
    drawFullEyeDiagram(x_phase/max(x_phase),Fs=fs,T=1)
    plt.title('in phase')
    plt.subplot(1,2,2)
    drawFullEyeDiagram(x_qua/max(x_qua),Fs=fs,T=1)
    plt.title('in quadrature')
    
```

```python
num_symbols = [4,16,64]

plt.figure(figsize=(20,5))
for i,num_simbol in enumerate(num_symbols):
    plt.subplot(1,3,i+1)
    QAM = komm.QAModulation(num_simbol)
    plt.plot(QAM.constellation.real,QAM.constellation.imag, 'r*',markersize=12)
    plt.grid()
    plt.title('{}-QAM'.format(num_simbol))
    plt.xlabel('I')
    plt.ylabel('Q')
```

### Ejemplo de todos los simbolos posibles que puede tener 16-QAM

```python
all_bits = combin(4).reshape(1,-1)[0]
QAM_16 = komm.QAModulation(16)

A = QAM_16.modulate(all_bits)

t = np.arange(0,1,1/fs)
x = np.cos(2*np.pi*t*1)
y = np.sin(2*np.pi*t*1)

plt.figure(figsize=(20,6))
for i in A:
    plt.plot(t,i.real*x-i.imag*y)
```

### Ejemplo de secuencia de bits modulada con 16-QAM

```python
bits = np.random.randint(0,2,20)
tB = 1 #tiempo de bit 
fc = 1 #frecuencia de la portadora 

t,bits_t = x_t(bits,tB,0)
plt.figure(figsize=(20,5))
plt.plot(t,bits_t)
for j in range(len(bits)):
    plt.axvline(j*tB, color='r', ls="dotted")
plt.title('secuencia de bits')
    
    
symbols= QAM_16.modulate(bits)
tS = 1 #timepo de simbolo 
t,xt = x_t(symbols,tS,fc)

plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
plt.plot(t,xt)
for j in range(len(symbols)):
    plt.axvline(j*tS, color='r', ls="dotted")
plt.title('secuencia modulada 16-QAM')
plt.subplot(1,2,2)
F = np.fft.fft(xt) 
f_ = np.fft.fftfreq(len(F),1/fs)
plt.plot(f_[range(100)],abs(F[range(100)]))
plt.title("Espectro")

span = 0.1
roll_off = 0.5

h = raisedCosinePulse(roll_off,span)*np.cos(2*np.pi*np.arange(-4*span,4*span,1/fs)*fc)
x_send =  np.convolve(xt,h,mode="same")
x_send = x_send/max(x_send)
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
plt.plot(t,x_send)
for j in range(len(symbols)):
    plt.axvline(j*tB, color='r', ls="dotted")
plt.title("Resultado del RCC")
plt.subplot(1,2,2)
F = np.fft.fft(x_send) 
f_ = np.fft.fftfreq(len(F),1/fs)
plt.plot(f_[range(100)],abs(F[range(100)]))
plt.title("Espectro")

eye_diagram(QAM_16,100,h)
```

## Distacia Hamming 

```python
group_bits= combin(4)
group_bits
```

```python
def distancia_hamming(group_bits):
    distacias = np.zeros((group_bits.shape[0],group_bits.shape[0]))
    for i,bits_i in enumerate(group_bits): 
        for j,bits_j in enumerate(group_bits):
            distacias[i,j] = np.sum(bits_i!=bits_j)
    return distacias
```

```python
dist = distancia_hamming(group_bits)
dist
```

```python
row_labels = list([str(x) for x in group_bits])
column_labels = list([str(x) for x in group_bits])

fig, ax = plt.subplots(figsize=(20,5))

heatmap = ax.pcolor(dist, cmap=plt.cm.seismic)

fig.colorbar(heatmap)

ax.set_xticks(np.arange(dist.shape[1])+0.5, minor=False)
ax.set_yticks(np.arange(dist.shape[0])+0.5, minor=False)

ax.invert_yaxis()
ax.xaxis.tick_top()

ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(column_labels, minor=False)
plt.show()
```

### BER VS SNR

```python
bits_TX = np.random.randint(0,2,10000) 
SNR = np.arange(0.1,5,0.1)
QAM_16 = komm.QAModulation(16) 

error = []
for i in SNR:
    simbols_TX = QAM_16.modulate(bits_TX)
    canal = komm.AWGNChannel(i)
    simbols_RX = canal(simbols_TX)
    bits_RX = QAM_16.demodulate(simbols_RX)
    error.append(np.sum(bits_RX != bits_TX))
```

```python
plt.figure(figsize=(20,6))

plt.subplot(1,2,1)
plt.stem(SNR,error, use_line_collection = True)
plt.xlabel('SNR')
plt.ylabel('BER')

plt.subplot(1,2,2)
etiquetas = QAM_16.bits_to_symbols(bits_TX)
for i in range(QAM_16.order):
    plt.plot(simbols_RX[etiquetas == i].real,simbols_RX[etiquetas == i].imag,'xk')
plt.plot(QAM_16.constellation.real,QAM_16.constellation.imag, 'r*',markersize=12)
plt.title('Max SNR')
plt.xlabel('I')
plt.ylabel('Q')
```

* **Codificación y distancia Hamming**: 

  1. **Codificación Hamming**: Es una codificación encaminada a la detección y corrección de errores que se puede presentar en la trama. Su funciomamiento se basa en agragar unos bits que se conocen como **“bits de redundancia”** pues son bits extra a la información para intentar corregir los errores o al menos comprobar que todo esté correcto. El numero de bits extra puede ser calculado con la siguiente formula 
  $$2^r \geq m + r + 1$$
  donde $r$ = numero de bits redundates, $m$ longitud de los datos 
  El código Hamming usa los **bits de paridad**, 1 es un número de unos el los datos igual a un número impar, mientras el 0 significa que existe un número para de unos en los datos. \
  **Algoritmo** 
     1. Escriba las posiciones de bit comenzando desde 1 en forma binaria (1, 10, 11, 100, etc.).
     2. Todas las posiciones de bit que tienen una potencia de 2 están marcadas como bits de paridad (1, 2, 4, 8, etc.)
     3. Todas las otras posiciones de bits están marcadas como bits de datos. 
     4. Cada bit de datos se incluye en un conjunto único de bits de paridad, según se determina su posición de bit en forma binaria.
         1. El bit de paridad 1 cubre todas las posiciones de bits cuya representación binaria incluye un 1 en el menos significativo
posición (1, 3, 5, 7, 9, 11, etc.).
         2. El bit de paridad 2 cubre todas las posiciones de bits cuya representación binaria incluye un 1 en la segunda posición desde
el bit menos significativo (2, 3, 6, 7, 10, 11, etc.).
         3. El bit de paridad 4 cubre todas las posiciones de bits cuya representación binaria incluye un 1 en la tercera posición desde
el bit menos significativo (4–7, 12–15, 20–23, etc.).
         4. El bit de paridad 8 cubre todas las posiciones de bits cuya representación binaria incluye un 1 en la cuarta posición desde
los bits de bit menos significativos (8–15, 24–31, 40–47, etc.)
         5. En general, cada bit de paridad cubre todos los bits donde el Y bit a bit de la posición de paridad y la posición de bit es distinto de cero
     5. Dado que verificamos la paridad par, establecemos un bit de paridad en 1 si el número total de unidades en las posiciones que verifica es
impar.
     6. Establezca un bit de paridad en 0 si el número total de unidades en las posiciones que verifica es par.
     7. **Decodificando un mensaje en Hamming Code**
        1. Cálculo del número de bits redundantes.
        2. Posicionamiento de los bits redundantes.
        3. Comprobación de paridad.
        4. Detección y corrección de errores.
     
    **Ejemplo**
    Queremos enviar la secuencia 1011001
    M = 7, r = 4, el total es 11, la secuencia con los bits de paridad seriá 
    $101R_8100R_41R_2R_1$
    Mirando las posiciones los valores de paridad serian 
    $R_1 = 0, R_2 = 1, R_4 = 1, R_8 = 0 $
    Entoces la secuencia seria 10101001110 \
    Ahora para detectar y corrgir errores debemos cer la paridad de los poscicines que le corespondieron a cada R, por ejemplo supongamos que la secuenia recivida tiene un error 10101101110 para $R_1 = 0, R_2 = 1, R_4 = 1,R_8 = 0 $
    con ello tenemos que hay un error en la posicion 0110 = 6, si no hubiera error daria 0000
    
  2. **Distancia Hamming**: La distanciá Hamming es una distacia que se mide con el número de bits diferentes que tiene dos palabras de bits, ejemplo 000 y 101 la distancia Hamming entre ellos es 2.  


* **Curva de Errores de Bit vs. SNR** : La curva BER vs. SNR es una gráfica donde podemos ver el número de errores que se presentan en la recepción del mensaje según un valor especifico de SNR presente en el sistema de comunicación. Un ejemplo es presentado en la sección [BER VS SNR](#BER-VS-SNR)

**Metodológica de Diseño**
1. Genere los bits aleatoriamente y agrúpelos en “Palabras de Bit” de tamaño igual a potencias de dos (2, 4, 8, etc.)
2. Determine el número de “Símbolos” con la ecuación: Número de Símbolos = (Palabras de Bit) 2
3. Asigne cada “Palabra de Bit” a un “Símbolo” (mapping and coding).
4. Ubique los “Símbolos” en el diagrama de constelación.
5. Proyecte los  ́”Símbolos” al eje “I” y determine el número de niveles de la señal PAM – I.
6. Proyecte los “Símbolos” al eje “Q” y determine el número de niveles de la señal PAM – Q.
7. Diseño el filtro RRC y genere los pulsos (pulse shaping) para las señales PAM – I y PAM – Q.(caracterice las señales PAM-I y PAM-Q con el diagrama de ojo)

comaparación entre ambos sistemas 

```python
def plot_ber(mod):
    bits_TX = np.random.randint(0,2,10000) 
    SNR = np.arange(0.1,5,0.1)
    

    error = []
    for i in SNR:
        simbols_TX = mod.modulate(bits_TX)
        canal = komm.AWGNChannel(i)
        simbols_RX = canal(simbols_TX)
        bits_RX = mod.demodulate(simbols_RX)
        error.append(np.sum(bits_RX != bits_TX))


    plt.figure(figsize=(20,6))

    plt.subplot(1,2,1)
    plt.stem(SNR,error, use_line_collection = True)
    plt.xlabel('SNR')
    plt.ylabel('BER')

    plt.subplot(1,2,2)
    etiquetas = mod.bits_to_symbols(bits_TX)
    for i in range(mod.order):
        plt.plot(simbols_RX[etiquetas == i].real,simbols_RX[etiquetas == i].imag,'x')
    plt.plot(mod.constellation.real,mod.constellation.imag, 'k*',markersize=12)
    plt.title('Max SNR')
    plt.xlabel('I')
    plt.ylabel('Q')
```

```python
plot_ber(espiral_16)
plot_ber(star_16)
```

* **Proponga una codificación que maximice la distancia Hamming (slide 45)**
Se diseña una codificación que maximece la distancia Hamming para la mudalación 16-QAM. Lo que se pretende hacer es hacer un grupo por cuadrante, como tenemos 4 cuadrantes entonces tenemso 4 grupos de 4 simbolos cada uno,como se puede ver en la sección [Distacia Hamming ](#Distacia-Hamming) en la gráfica de distancias existe cierta simetria, entonces aprovechando empezamos escogiendo la paraja de simbolos que tenga mayor distancia entre si, que seria las parajas que están pintadas de rojo mas intenso, con ello obtenemos 8 grupos, seguidamente necesitamos juntamos las dos secuencias de bits y volvemos a montar las distancias entre los grupos, y realizamos el mismo porcedimiento agrupamos los que tengan mayor distancia. 

```python
group_bits= combin(4)
```

```python
dist = distancia_hamming(group_bits)[:,:8]
dist
```

```python
cor = np.where(dist == 4)
cor
```

primeros 8 grupos 

```python
new = np.concatenate((group_bits[cor[0]],group_bits[cor[1]]),axis =1)
new
```

```python
dist_2 = distancia_hamming(new)[:,:4]
dist_2
```

```python
cor_2 = np.where(dist_2 == 6)
cor_2
```

```python
new_2 = np.concatenate((new[cor_2[0]],new[cor_2[1]]),axis =1)
new_2
```
