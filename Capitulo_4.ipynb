### Defina:
* **Interferencia Inter-Canal** : La interferencia inter-canal se produce cuando los componentes espectrales de  canales adyacentes se mezclan uno con el otro, produciendo distorción en las señales. Este fenomeno se puede presentar cuando no se realiza un adecuado filtrado, o no se sintoniza bien el canal. Una forma de solucionarlo ha sido la implementación de las llamadas bandas de guarda, que son una separación de los canales en el espectro. 

![](./images/banda.png)
* **Multiplexación en el dominio de la frecuencia** : Como se sabe el concepto básico de multiplexar es usar un mismo canal para varios usuarios a la vez, entonces como en la multiplexación en el tiempo donde a cada usuario le damos un espacio de tiempo para transmitir, en la frecuencia le damos al usuario un rango de frecuencias que debe estar dentro del las frecuencias admisibles por medio de transmición, en la siguiente imagen se ilustra este concepto  

![](./images/FDM.png)


* **Esquema de transmisión Gridless**: Este esquema se basa en que todos los canales no necitan igual ancho de banda todo el tiempo, en otras palabras tiene en cuenta la dinámica de cada canal al requerir ancho de banda, con ello en mente no tiene rangos de frecuencia fijos para cada canal, si no que van mutando, incrementado o decrementando, según el canal lo requiera. En la siguiente imagen se puede ver una comparación de Fized Grid y GridLess en fibra optica. 

![](./images/grid.jpg)

#### Capacidad del Canal 
La capacidad de un canal la podemos medir de la siguiente forma
Según Nyquist:

$$C = 2B\log_2{N}\ \ [bps]$$

Según Shannon:

$$C = B\log_2(1 + SNR)\ \ [bps]$$

donde $B$ es el nacho de banda analogico, $N$ es el numéro de simbolos del esquema de modulación y $SNR$ es la relación señal a ruido. 

Para el ejercicio vamos a utilizar un valore de $SNR_{dB} = 25 \ dB$ lo que sería igual a $SNR = 316.22$

Con lo que tendriamos aproximadamente 

$$C = 8B \ \ [bps]$$

```python
import numpy as np 
import matplotlib.pyplot as plt 
import komm

fs = 10000 #frecuencia "real" de muestreo 
```

```python

def x_t(simbols,tS,fc=1,separadas=False):
    """tS timepo de duración del simbolo"""
    t_c = np.arange(0,tS,1/fs)
    
    tf = tS*len(simbols)
    t = np.arange(0,tf,1/fs)

    x = np.cos(2*np.pi*t_c*fc)
    y = np.sin(2*np.pi*t_c*fc)
    
    if not separadas:
        fm = np.array([])
        for i in simbols:
            fm = np.append(fm, i.real*x-i.imag*y) 
        return t,fm[:len(t)]
    
    else: 
        phase = np.array([])
        qua = np.array([])
        for i in simbols:
            phase = np.append(phase, i.real*x) 
            qua = np.append(qua,i.imag*y) 
        return phase,qua
    
    
def raisedCosinePulse(t,roll_off=0.0,span=1):
    return np.sinc(t/span)*np.cos(np.pi*roll_off*t/span)/(1-(4*roll_off**2*t**2)/span**2)

def rootRaisedCosinePulse(t,roll_off=1,span=1):
    a = (2*roll_off)/(np.pi*np.sqrt(span))
    b = (np.cos((1+roll_off)*np.pi*t/span)+ np.sin((1-roll_off)*np.pi*t/span)/(4*roll_off*t/span ))
    return a*b/(1-(4*roll_off*t/span)**2)
```

```python
bits = np.random.randint(0,2,256)
```

```python
canal_1 = komm.QAModulation(16)
simbols = canal_1.modulate(bits)
C1t_,C1xt = x_t(simbols,tS=1./15.,fc=25,separadas=False)

canal_2 = komm.QAModulation(16)
simbols = canal_2.modulate(bits)
C2t_,C2xt = x_t(simbols,tS=1./20.,fc=61,separadas=False)

canal_3 = komm.PSKModulation(4,phase_offset=np.pi/4)
simbols = canal_3.modulate(bits)
C3t_,C3xt = x_t(simbols,tS=1./20.,fc=92,separadas=False)
len(C1xt)
```

```python
roll_offs = [0.99,0.99,0.0001]
channels = [C1xt,C2xt,C3xt]
f_carriers = [25,61,92]
for xt in channels:
    F = np.fft.fft(xt) 
    f_ = np.fft.fftfreq(len(F),1/fs)
    plt.plot(f_[range(int(len(F)/70))],abs(F[range(int(len(F)/70))])/abs(max(F)))
    plt.title("Espectro Normalizado")
    
plt.legend(['Canal 1','Canal 2','Canal 3'])

for fc in f_carriers:
    plt.axvline(fc, color='k', ls="dotted")

plt.axvline(10, color='r', ls="dotted")
plt.axvline(110, color='r', ls="dotted")
```

```python
plt.figure(figsize=(20,10))
for t_,xt,roll_off,fc in zip([C1t_,C2t_,C3t_],channels,roll_offs,f_carriers):
    impulse  = raisedCosinePulse(t_,roll_off)*np.cos(t_*2*np.pi*fc)
    output = np.convolve(impulse,xt, mode="same")
    F = np.fft.fft(output) 
    f_ = np.fft.fftfreq(len(F),1/fs)
    
    plt.plot(f_[range(int(len(F)/70))],abs(F[range(int(len(F)/70))])/abs(max(F)))


plt.title("Espectro")    
plt.legend(['Canal 1','Canal 2','Canal 3'])

plt.axvline(10, color='r', ls="dotted")
plt.axvline(110, color='r', ls="dotted")

plt.axvline(10, color='b', ls="dotted")
plt.axvline(40, color='b', ls="dotted")

plt.axvline(41, color='orange', ls="dotted")
plt.axvline(81, color='orange', ls="dotted")

plt.axvline(82, color='g', ls="dotted")
plt.axvline(102, color='g', ls="dotted")

```
